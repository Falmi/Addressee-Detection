```python
from PIL import Image

def imcrop(image_path, coords, saved_location):
    #coords=coords.split(", ").replace('[','').replace(']','')
    li = list(coords.split(", "))
    cord=[]
    for i in li:
        cord.append(int(i.replace('[','').replace(']','')))

    image_obj = Image.open(image_path)
    cropped_image = image_obj.crop((cord[0],cord[1],cord[0]+cord[2],cord[1]+cord[3]))
    cropped_image.save(saved_location)
```

```python
import os, cv2, glob, numpy, random


datapath ="./data/croped_face_region/"
numFrames=9
VisualAug= True
csv_path="data/E_MUMMER_csv/all_E_MUMMER_dataset.csv"

mixLst = open(csv_path).read().splitlines()
data=mixLst[0]
def load_visual(data, dataPath, numFrames, visualAug): 

 
    row=data.split("\t")
    image_paths=[]
    if int(row[1])==1:
        image_path.append(f"{row[0]}.jpg")
    else:
        start_frm=round(float(row[0].split("_")[-2])*15)
        end_frm=round(float(row[0].split("_")[-1])*15)
        strt=start_frm
        fl_names=row[0].split(":")
        subj_id=fl_names[1].split("_")[0]
        datapath ="./data/croped_face_region/"

        while start_frm <=end_frm:
            print(round(start_frm/15,2))
            frm_ts =round(start_frm/15,2)
            fl_name=f"{fl_names[0]}:{subj_id}_{frm_ts}.jpg"
            image_path=os.path.join(datapath,fl_name)
            image_paths.append(image_path)
            print(image_path)
            start_frm+=1
    
    #print(data.split('/')[-1][:-4])
    #sortedFaceFiles = sorted(image_paths, key=lambda data: (float(data.split('/')[-1][:-4])), reverse=False) 
    faces = []
    H = 112
    
    if visualAug == True:
        new = int(H*random.uniform(0.7, 1))
        x, y = numpy.random.randint(0, H - new), numpy.random.randint(0, H - new)
        M = cv2.getRotationMatrix2D((H/2,H/2), random.uniform(-15, 15), 1)
        augType = random.choice(['orig', 'flip', 'crop', 'rotate']) 
    else:
        augType = 'orig'
    for faceFile in image_paths[:numFrames]:
        face = cv2.imread(faceFile)
        face = cv2.cvtColor(face, cv2.COLOR_BGR2GRAY)
        face = cv2.resize(face, (H,H))
        if augType == 'orig':
            faces.append(face)
        elif augType == 'flip':
            faces.append(cv2.flip(face, 1))
        elif augType == 'crop':
            faces.append(cv2.resize(face[y:y+new, x:x+new] , (H,H))) 
        elif augType == 'rotate':
            faces.append(cv2.warpAffine(face, M, (H,H)))
    faces = numpy.array(faces)
    return faces
```
